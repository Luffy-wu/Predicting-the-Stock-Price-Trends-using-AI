# Plot (Standardized Data)

```python
def standize(df):
    df_norm=df.sub(df.mean(axis=1),axis=0).divide(df.std(axis=1),axis=0)
    return df_norm
df_close=pd.read_pickle('close')[df_returns_filter.columns]
df_high=pd.read_pickle('high')[df_returns_filter.columns]
df_low=pd.read_pickle('low')[df_returns_filter.columns]
df_volume=pd.read_pickle('volume')[df_returns_filter.columns]
df_open=pd.read_pickle('open')[df_returns_filter.columns]
df_close=standize(df_close)
df_high=standize(df_high)
df_low=standize(df_low)
df_volume=standize(df_volume)
df_open=standize(df_open)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16,10))
fig.subplots_adjust(hspace=0.4)
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(233)
ax4 = fig.add_subplot(234)
ax5 = fig.add_subplot(235)
df_close.plot(ax=ax1,legend=False,title='Close')
df_high.plot(ax=ax2,legend=False,title='High')
df_low.plot(ax=ax3,legend=False,title='Low')
df_open.plot(ax=ax4,legend=False,title='Open')
df_volume.plot(ax=ax5,legend=False,title='Volume')
plt.show()
```
