```python
import talib
import pandas as pd,numpy as np
```

```python
share1=get_price('000001.XSHE',start_date='2015-01-01',end_date='2016-01-01')
close=share1['close'].values
high=share1['high'].values
low=share1['low'].values
volume=share1['volume'].values
open=share1['open'].values

RSI=talib.RSI(close,timeperiod=14)
macd, macdsignal, macdhist = talib.MACD(close, fastperiod=12, slowperiod=26, signalperiod=9)
CCI = talib.CCI(high, low, close, timeperiod=14)
CMO = talib.CMO(close, timeperiod=14)
ATR = talib.ATR(high, low, close, timeperiod=14)
MA5 = talib.MA(close, timeperiod=5, matype=0)
MA10 = talib.MA(close, timeperiod=10, matype=0)
MA=MA5/MA10
```

```python
df_temp=np.array([RSI,macd,CCI,CMO,-ATR,MA]).reshape(6,-1).T
df=pd.DataFrame(df_temp,index=share1.index,columns=['RSI','MACD','CCI','CMO','ATR','MA'])
returns=[0]
returns.extend(list(close[1:]/close[:-1]-1))
df['returns']=returns
df
```
